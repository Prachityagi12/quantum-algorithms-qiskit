```python
import random
import string
from qiskit import QuantumCircuit,transpile
from qiskit_aer import Aer
from qiskit.circuit.library import GroverOperator, MCMT, ZGate

import numpy as np

# Set parameters
string_length = 10
trials = 100

# Initialize probabilities
prob_a, prob_b, prob_c, prob_d = 0, 0, 0, 0

# Simulation loop
for _ in range(trials):
    # Generate a random string
    random_str = ''.join(random.choices(string.ascii_lowercase, k=string_length))

    # Quantum setup
    qubits = len(random_str)
    search_letter = 'a'  # Change dynamically per trial

    # Oracle creation
    oracle_circ = QuantumCircuit(qubits)
    marked_indices = [idx for idx, char in enumerate(random_str) if char == search_letter]
    if marked_indices:
        oracle_circ.x(marked_indices)
        oracle_circ.compose(MCMT(ZGate(), qubits - 1, 1), inplace=True)
        oracle_circ.x(marked_indices)

    # Grover's operator
    grover = GroverOperator(oracle_circ)

    # Optimal number of iterations
    iterations_needed = int(np.floor(np.pi / 4 * np.sqrt(2 ** qubits)))

    # Build quantum circuit
    quantum_circ = QuantumCircuit(qubits, qubits)
    quantum_circ.h(range(qubits))
    quantum_circ.compose(grover.power(iterations_needed), inplace=True)
    for q in range(qubits // 2):
        quantum_circ.swap(q, qubits - q - 1)
    quantum_circ.measure_all()

    # Run simulation
    simulator_backend = Aer.get_backend('qasm_simulator')
    compiled = transpile(quantum_circ, simulator_backend)
    sim_result = simulator_backend.run(compiled, shots=5).result()
    measurement_results = sim_result.get_counts()
    highest_prob_state = max(measurement_results, key=measurement_results.get)
    count_of_ones = highest_prob_state.count('1')
    prob_a += count_of_ones / qubits

    # Repeat for 'b'
    search_letter = 'b'
    oracle_circ = QuantumCircuit(qubits)
    marked_indices = [idx for idx, char in enumerate(random_str) if char == search_letter]
    if marked_indices:
        oracle_circ.x(marked_indices)
        oracle_circ.compose(MCMT(ZGate(), qubits - 1, 1), inplace=True)
        oracle_circ.x(marked_indices)
    grover = GroverOperator(oracle_circ)
    quantum_circ = QuantumCircuit(qubits, qubits)
    quantum_circ.h(range(qubits))
    quantum_circ.compose(grover.power(iterations_needed), inplace=True)
    quantum_circ.measure_all()
    compiled = transpile(quantum_circ, simulator_backend)
    sim_result = simulator_backend.run(compiled, shots=5).result()
    measurement_results = sim_result.get_counts()
    highest_prob_state = max(measurement_results, key=measurement_results.get)
    count_of_ones = highest_prob_state.count('1')
    prob_b += count_of_ones / qubits

    # Repeat for 'c'
    search_letter = 'c'
    oracle_circ = QuantumCircuit(qubits)
    marked_indices = [idx for idx, char in enumerate(random_str) if char == search_letter]
    if marked_indices:
        oracle_circ.x(marked_indices)
        oracle_circ.compose(MCMT(ZGate(), qubits - 1, 1), inplace=True)
        oracle_circ.x(marked_indices)
    grover = GroverOperator(oracle_circ)
    quantum_circ = QuantumCircuit(qubits, qubits)
    quantum_circ.h(range(qubits))
    quantum_circ.compose(grover.power(iterations_needed), inplace=True)
    quantum_circ.measure_all()
    compiled = transpile(quantum_circ, simulator_backend)
    sim_result = simulator_backend.run(compiled, shots=5).result()
    measurement_results = sim_result.get_counts()
    highest_prob_state = max(measurement_results, key=measurement_results.get)
    count_of_ones = highest_prob_state.count('1')
    prob_c += count_of_ones / qubits

    # Repeat for 'd'
    search_letter = 'd'
    oracle_circ = QuantumCircuit(qubits)
    marked_indices = [idx for idx, char in enumerate(random_str) if char == search_letter]
    if marked_indices:
        oracle_circ.x(marked_indices)
        oracle_circ.compose(MCMT(ZGate(), qubits - 1, 1), inplace=True)
        oracle_circ.x(marked_indices)
    grover = GroverOperator(oracle_circ)
    quantum_circ = QuantumCircuit(qubits, qubits)
    quantum_circ.h(range(qubits))
    quantum_circ.compose(grover.power(iterations_needed), inplace=True)
    quantum_circ.measure_all()
    compiled = transpile(quantum_circ, simulator_backend)
    sim_result = simulator_backend.run(compiled, shots=5).result()
    measurement_results = sim_result.get_counts()
    highest_prob_state = max(measurement_results, key=measurement_results.get)
    count_of_ones = highest_prob_state.count('1')
    prob_d += count_of_ones / qubits

# Calculate and print probabilities
print("Probability of occurrence of 'a':", prob_a / trials)
print("Probability of occurrence of 'b':", prob_b / trials)
print("Probability of occurrence of 'c':", prob_c / trials)
print("Probability of occurrence of 'd':", prob_d / trials)
```
